## Ejemplo 2: Manipulación de Strings

### 1. Objetivos:
    - Aprender a usar `replace`, `strip`, `title`, `upper`, `lower` y `split` para transformar datos tipo `string``
 
---
    
### 2. Desarrollo:

```python
import pandas as pd
```

```python
df = pd.read_csv('../Datasets/new_york_times_bestsellers-dirty.csv', index_col=0)
df.head(3)
```

Empecemos con la columna `description` que tiene un 'Descr:' al inicio de cada texto. Si queremos remover ese texto podemos usar el método `replace` de la propiedad `str` de esa `Serie` en la forma:

`dataframe[-columna-].str.replace("origen", "destino")`

```python
df["description"] = df["description"].str.replace(
    "Descr:", "")
df.head(3)
```

Podemos consultas una fila de la columna `description` para comprobar el resultado con:

`dataframe.loc[-fila-, -columna-]`

```python
df.loc[0, "description"]
```

Como puedes ver, tenemos también espacios vacíos al principio y final de nuestras `strings`. Vamos a removerlos usando `strip` en la forma:

`dataframe[-columna-].str.strip()`

```python
df['description'] = df['description'].str.strip()
df.loc[0, "description"]
```

Perfecto.

Ahora veamos la columna 'title', cuyos textos están en mayúsculas. Esto no es muy agradable, así que podemos usar algunos métodos para modificar el patrón de mayúsculas y minúsculas:

```python
# titulos a minúsculas
df["title"].str.lower()
```

```python
# títulos como nombres
df["title"].str.title()
```

Este último es más adecuado, vamos a guardarlo:

```python
# remplazar la columna
df["title"] = df["title"].str.title()
df.head()
```

Ahora, digamos que queremos separar nuestra columna `author` en dos columnas `author_first_name` y `author_last_name`. Eso lo podemos hacer con el método `split` en la forma:

`dataframe[-columna-].str.split(-separador-)`

```python
df.loc[0, "author"]
```

```python
df["author"].str.split(" ")
```

Podemos convertirlo en dos columnas así:

`dataframe[-columna-].str.split(-separador-, expand=True)`

```python
df["author"].str.split(" ", expand=True)
```

```python
# asignando las nuevas columnas
df[ ["author_first_name", "author_last_name"] ] = \
    df["author"].str.split(" ", expand=True)
```

```python
# revisando el resultado con head()
df.head(3)
```
