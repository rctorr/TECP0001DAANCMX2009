## Insertando dantos en una base de datos SQL

```python
import sqlite3
```

```python
conn = sqlite3.connect("Datasets/movielens.sqlite3")
cur = conn.cursor()
cur
```

```python
sql = "SELECT * FROM users LIMIT 5"
cur.execute(sql)
```

```python
resul = cur.fetchall()
resul
```

```python
fila_1 = (1, "F", 25, 7, "66666")
```

```python
sql = "INSERT INTO users VALUES(?,?,?,?,?)"
cur.execute(sql, fila_1)
conn.commit()
```

```python
filas = [
    (1, "F", 25, 7, "66666"),
    (2, "M", 56, 1, "99999")
]
```

```python
# forma super eficiente
sql = "INSERT INTO users VALUES(?,?,?,?,?)"
for x in filas:
    cur.execute(sql, x)
    
conn.commit()   # <- razÃ³n
```
